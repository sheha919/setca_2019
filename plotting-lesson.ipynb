## Plotting Lesson

```python
import numpy
import os
```

```python
distance_file=os.path.join('data', 'distance_data_headers.csv')
distances = numpy.genfromtxt(fname=distance_file, delimiter=',' ,dtype='unicode')
print(distances)
```

```python
headers= distances[0]
data= distances[1:]
data = data.astype(numpy.float) #changing the str to float data type 
```

```python
print(headers)
print(data)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure()
plt.plot(data[:,1])
```

```python
plt.figure()
plt.plot(data[:,0],data[:,1])
```

```python
sample =headers[1]

plt.figure()
plt.xlabel('Simulation Frame')
plt.ylabel('Distance (Angstroms)')
fig_1 =plt.plot(data[:,0],data[:,1], label=sample)
plt.legend()
plt.savefig(F'{sample}.png') #export and save the plot
```

```python
#multiple plots
plt.figure()
plt.xlabel('Simulation Frame')
plt.ylabel('Distance (Angstroms)')
plt.plot(data[:,0],data[:,1], label=headers[1])
plt.plot(data[:,0],data[:,2], label=headers[2])
plt.legend()
```

```python
#plot every column using for loop
plt.figure()
for col in range(1, len(headers)):
    fig =plt.plot(data[:,0],data[:,col], label=headers[col])
plt.xlabel('Simulation Frame')
plt.ylabel('Distance (Angstroms)')
plt.legend()
plt.savefig('all_samples_300dpi.png', dpi=300) #resolution can be enhance by dpi number
```

```python
#modify for loop to change the above graph into seperate plots
for col in range(1, len(headers)):
    plt.figure()
    fig =plt.plot(data[:,0], data[:,col], label=headers[col])
    plt.xlabel('Simulation Frame')
    plt.ylabel('Distance (Angstroms)')
    plt.legend()
    plt.savefig('all_samples[col].png', dpi=300)
```

```python
#print rows 0 to 10 and skip by 2 with all columns
print(data[0:10:2,:])
```
